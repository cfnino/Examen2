````python
# Importing required libraries
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from IPython.core.display import HTML
````

````python
# Base URL of the site to be analyzed
SITE_URL = 'https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia'
````

````python
# Firefox web driver path
# Download the driver for you S.O. here: https://github.com/mozilla/geckodriver/releases
DRIVER_PATH = './geckodriver.exe'
````

````python
# Creating a new firefox window
browser = webdriver.Firefox(executable_path = DRIVER_PATH)
````

Obtenemos el contenido HTML de la página:

````python
browser.get(SITE_URL)
soup = BeautifulSoup(browser.page_source, 'html.parser')
````

Analizando el HTML de la página encontramos que las tablas de Wikipedia para esta página son así:



En esta parte buscamos el tag de tipo table, y que su clase se llame wikitable:

````python
tables = soup.find_all('table', class_ = 'wikitable')
````

Creamos un arreglo llamado blocks para poder buscar los tags de tr dentro de cada resultado de tables.

````python
blocks = []
for i in tables:
    blocks.append(i.find_all('tr'))

print("Cantidad de tablas: "+ str(len(blocks)))
````

Podemos observar la cantidad encontrada de tags tr por cada tabla, esto haría referencia a cada presidente (tr referencia a las filas de la tabla).

**Nota:** El primer tr de cada tabla no contiene información.

````python
print("Cantidad de tr por tabla: ")
for b in range(0,len(blocks)):
    t = b+1
    print(str(t)+'-'+str(len(blocks[b])))
````

Analizando las tablas se observa que todos contienen las mismas columnas excepto la primera tabla de todas, que contiene una columna demás, por esta razón sacamos los títulos de la primera tabla para trabajar con estos títulos en general.

Para esta parte buscamos los tag th que hacen referencia a los títulos.

````python
headers = blocks[0][0].find_all('th')
titles = []
for h in headers:
    title = h.text.strip()
    titles.append(title)
titles
````

En esta parte vamos a recorrer todos los elementos de blocks, que es la lista de tags de tr, y vamos a buscar los td que sería cada columna de esa fila, almacenando el resultado en una variable content.

Luego vamos a recorrer content, para poder obtener los datos y guardarlos en una lista llamada con, esta lista hace referencia a un presidente.

Tenemos otra lista llamada cons, para poder guardar cada presidente.

````python
cons = []
for b in range(0, len(blocks)):
    for r in range(1,len(blocks[b])):
        content = blocks[b][r].find_all('td')
        con = []
        for c in range(0, len(content)):
            if(c != 0):
                # Como estamos utilizando los títulos de la primera tabla, significa que nos falta
                # una columna por rellenar para las siguientes tablas, por esta razón validamos que estemos 
                # en una tabla diferente a la primera y agregamos un campo vacío antes de la columna 4 que haría referencia al título de ciudades.
                if(c == 4 and b!=0):
                    con.append('')
                    con.append(content[c].text.strip())
                else:
                    # Para la columna dos está la imagen del presidente por esta razón volvemos a buscar un tag
                    # que hace referencia a la imagen, el tag que buscamos es img y obtenemos el valor src de este tag
                    if(c == 2):
                        img = content[c].find_all('img')
                        # Validamos si tiene información
                        if(len(img) > 0):
                            con.append(img[0]['src'])
                        else:
                            con.append('')
                    else:
                        con.append(content[c].text.strip())
        cons.append(con)
# Creamos el dataframe con los datos capturados y los títulos.
df = pd.DataFrame(cons, columns=titles)
````

Para mostrar todas las filas del dataframe y mostrar toda la información de las columnas:

En esta parte movemos las columnas a la posición que deseamos y cambiamos el nombre de dos columnas.

````python
df = df[['Presidente','Efigie','Período','Alineación','Ciudades','Comentarios']]
df = df.rename(columns={'Efigie':'Foto','Alineación':'Partido político'})
df
````

Eliminamos la columna de ciudades ya que en las tablas de Wikipedia solo la primera contaba con esta información, pero las otras 5 tablas de presidentes no la tenían, por esta razón decidí eliminar esta columna.

````python
df2 = df.drop(['Ciudades'],axis=1)
df2
````
